```python
from sklearn.model_selection import train_test_split
ml_data = data.copy()
x = ml_data.drop(columns=["Concrete compressive strength(MPa, megapascals)"])
y = ml_data["Concrete compressive strength(MPa, megapascals)"]

## Training set and testsing set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print(f"training x: {x_train.shape} y: {y_train.shape}")
print(f"testing x: {x_test.shape} y: {y_test.shape}")
```

 Training a linear regression model

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)
y_predictions = lr.predict(x_test)
```

Model Evaluation

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
r2 = r2_score(y_test, y_predictions)
mse = mean_squared_error(y_test, y_predictions)
mae = mean_absolute_error(y_test, y_predictions)

print(f"R^2: {r2} \n MSE: {mse} \n MAE: {mae}")
```

- R^2: 0.627541605542902: this shos that our model is 62% accurate
- MSE: 95.97548435337708: 
- MAE: 7.745392872421345: This tells us that the value our model is likely to predict will be 7.7454 up or down the actual value
